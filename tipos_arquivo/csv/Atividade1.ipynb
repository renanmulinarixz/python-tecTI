Criar um sistema de biblioteca que salve a lista em um csv


```python
import csv
```

```python
livros_emprestados = [
    ["Dom Casmurro", "Ana Silva", "2026-04-10", "2026-04-20"],
    ["1984", "Carlos Souza", "2026-04-12", "2026-04-22"],
    ["O Pequeno Príncipe", "Mariana Lima", "2026-04-15", "2026-04-25"],
    ["Capitães da Areia", "João Pereira", "2026-04-18", "2026-04-28"]
]
```

```python
with open("livros_emprestados.csv", mode="w", newline="", encoding="utf-8") as arquivo:
    writer = csv.writer(arquivo)
    # Cabeçalho
    writer.writerow(["titulo", "pessoa", "data_emprestimo", "data_devolucao"])
    # Dados
    writer.writerows(livros_emprestados)

print("Arquivo CSV criado com sucesso!")
```

Usando Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(livros_emprestados, columns=["titulo", "pessoa", "data_emprestimo", "data_devolucao"])
df.to_csv("livros_emprestados_pandas.csv", index=False, encoding="utf-8")
```

Usando o input

```python
# Crio uma lista vazia para receber os dados do CSV
dados = []
qtd = int(input("Quantos livros deseja registrar? "))
for i in range(qtd):
    titulo = input("Digite o título do livro: ")
    pessoa = input("Digite o nome da pessoa: ")
    data_emprestimo = input("Digite a data de empréstimo (YYYY-MM-DD): ")
    data_devolucao = input("Digite a data de devolução (YYYY-MM-DD): ")
    dados.append([titulo, pessoa, data_emprestimo, data_devolucao])
    dados.append({
        "titulo": titulo,
        "pessoa": pessoa,
        "data_emprestimo": data_emprestimo,
        "data_devolucao": data_devolucao
    })
df = pd.DataFrame(dados)
df.to_csv("livros_emprestados_usuario.csv", index=False, encoding="utf8")
```

Usando datetime

```python
import datetime as dt
bibliteca = []

while True:
    titulo = input("Digite o título do livro (ou 'sair' para encerrar): ")
    if titulo.lower() == 'sair':
        break
    pessoa = input("Digite o nome da pessoa: ")
    data_emprestimo = dt.datetime.now().strftime("%Y-%m-%d")  # Define a data de empréstimo como a data atual
    data_devolucao = dt.datetime.now() + dt.timedelta(days=14)  # Define a data de devolução para 14 dias após o empréstimo
    
    livro = {
        "titulo": titulo,
        "pessoa": pessoa,
        "data_emprestimo": data_emprestimo,
        "data_devolucao": data_devolucao
    }
    
    bibliteca.append(livro)
df = pd.DataFrame(bibliteca)
df.to_csv("livros_emprestados_biblioteca.csv", index=False, encoding="utf8")
```
